```python
# -----------------------------------------------------------
# Digital Signature System using ECDSA + SHA-256
# -----------------------------------------------------------

# 1. Import necessary libraries
from ecdsa import SigningKey, SECP256k1
import hashlib

# 2. Create a digital transaction message
transaction_message = "Ahmed pays 100 coins to Ali"
print("Transaction Message:", transaction_message)

# 3. Generate ECDSA key pair (private + public key)
private_key = SigningKey.generate(curve=SECP256k1)
public_key = private_key.get_verifying_key()

# 4. Hash the message using SHA-256
message_hash = hashlib.sha256(transaction_message.encode()).digest()
print("SHA-256 Hash:", hashlib.sha256(transaction_message.encode()).hexdigest())

# 5. Sign the hash using the private key
signature = private_key.sign(message_hash)
print("Signature:", signature.hex())

# 6. Verify the signature with the public key
is_valid = public_key.verify(signature, message_hash)
print("Signature Verified:", is_valid)
```
